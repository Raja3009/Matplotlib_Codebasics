Using following static data for companies revenues,

1) Plot a bar chart that looks like below

2) Save this chart as jpg file and then pdf file with 5 inch of padding



```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
import numpy as np
xpos = np.arange(len(company))
```

```python
xpos
```

```python
plt.xticks(xpos,company)

plt.bar(xpos,revenue,label= 'Revenue')
plt.xlabel('Company')
plt.ylabel('Revenue(bin)')
plt.title('US Technology Stocks')

plt.legend(loc = 0)
```

```python
plt.xticks(xpos,company)

plt.bar(xpos,revenue,label= 'Revenue')
plt.xlabel('Company')
plt.ylabel('Revenue(bin)')
plt.title('US Technology Stocks')

plt.legend(loc = 0)
plt.savefig('Companyrevenue.png',pad_inches = 5)
```

```python
plt.xticks(xpos,company)

plt.bar(xpos,revenue,label= 'Revenue')
plt.xlabel('Company')
plt.ylabel('Revenue(bin)')
plt.title('US Technology Stocks')

plt.legend(loc = 0)
plt.savefig('Companyrevenue.pdf',pad_inches = 5)
```
