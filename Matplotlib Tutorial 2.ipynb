## Matplotlib Tutorial: pyplot.plot format strings and other options


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,'g+')
```

```python
plt.plot(x,y,'g--')
```

```python
plt.plot(x,y,'--r*')
```

```python
plt.plot(x,y,'rD')
```

```python
plt.plot(x,y,color = 'red',marker = 'D', linestyle = '')
```

```python
plt.plot(x,y,color = '#fc03e8',marker = 'D', linestyle = '') #hex code
```

```python
plt.plot(x,y,color = 'green',marker = '+', linestyle = '',markersize = 20) #marker size
```

```python
plt.plot(x,y,color = 'green',marker = '+', alpha = 0.3) #tranasparency(0-1) 0---transparent and 1 ---opaque
```
